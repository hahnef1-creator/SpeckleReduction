```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#open the triple correlation for both the x and y subplanes
triple_correlation_y = cv2.imread('C:\\Users\Emily/h075695_1a_01_tc.tif',0)
triple_correlation_x = cv2.imread('C:\\Users\Emily/h075695_1a_10_tc.tif',0)

#take Fourier transform of both images and shift twice to find bispectral subplanes
shift=np.fft.fftshift(triple_correlation_y)
f_y = np.fft.fft2(shift)
fshift_y = np.fft.fftshift(f_y)
bispectrum_y = fshift_y

shift2=np.fft.fftshift(triple_correlation_x)
f_x = np.fft.fft2(shift2)
fshift_x = np.fft.fftshift(f_x)
bispectrum_x = fshift_x

#separate bispectral subplanes into 2d arrays of real and imaginary numbers
bispectrum_y_imag=bispectrum_y.imag
bispectrum_y_real=bispectrum_y.real
bispectrum_x_imag=bispectrum_x.imag
bispectrum_x_real=bispectrum_x.real

#find the phase shift for each direction
phi_y=np.arctan(bispectrum_y_imag/bispectrum_y_real)
phi_x=np.arctan(bispectrum_x_imag/bispectrum_x_real)

#create function to store the intial phase from phi_y and phi_x 
def phase_object(phi_x,phi_y):
    #create empty array
    phase_object=np.zeros(phi_x.shape,dtype="float32")
    height_phase,width_phase = phase_object.shape
    #set starting point as the center of the image and store as 0
    n=128
    m=128
    phase_object[n,m]=0
    #gives horizontal phase relative to the center
    for x in range(int(width_phase/2)-1):
        phase_object[n,m+x+1]=phase_object[n,m+x]-phi_x[n,m+x] 
        phase_object[n,m-x-1]=phase_object[n,m-x]-phi_x[n,m-x]
    #gives vertical phase relative to the center
    for y in range(int(height_phase/2)-1):
        phase_object[n+y+1,m]=phase_object[n+y,m]-phi_y[n+y,m]
        phase_object[n-y-1,m]=phase_object[n-y,m]-phi_y[n-y,m]
    return phase_object

#find all horizontal arrays from the given inital phase
def final_phase_image1(x_y_phase_map,phi_x):
    image=x_y_phase_map
    height,width=image.shape
    #need horizonal arrays above and below origin point, initialize the starting row above and below origin
    n=129
    n2=127
    m=128
    #iterate to find arrays below origin
    for one in range(int(height/2)-2):
        n+=1
        #iterate for the left and right side of the arrays 
        for mult in range(int(height/2)-2):
            image[n,m+mult+1]=image[n,m+mult]-phi_x[n,m+mult]
            image[n,m-mult-1]=image[n,m-mult]-phi_x[n,m-mult]
    #iterate to find arrays above origin
    for one in range(int(height/2)-2):
        n2-=1
        #iterate for the left and right side of the arrays
        for mult in range(int(height/2)-2):
            image[n2,m+mult+1]=image[n2,m+mult]-phi_x[n2,m+mult]
            image[n2,m-mult-1]=image[n2,m-mult]-phi_x[n2,m-mult]        
    return image

#find all vertical arrays from the given inital phase
def final_phase_image2(x_y_phase_map,phi_y):
    image=x_y_phase_map
    height,width=image.shape
    #initialize parameter for starting column from left and right column
    n=128
    m=129
    m2=127
    #iterate to find arrays to the right of the origin
    for one in range(int(width/2)-2):
        m+=1
        #iterate to find bottom and top half of arrays
        for mult in range(int(width/2)-2):
            image[n+mult+1,m]=image[n+mult,m]-phi_y[n+mult,m]
            image[n-mult-1,m]=image[n-mult,m]-phi_y[n-mult,m]
    #iterate to find arrays to the left of the origin
    for one in range(int(width/2)-2):
        m2-=1
        #iterate to find bottom and top half of arrays 
        for mult in range(int(width/2)-2):
            image[n+mult+1,m2]=image[n+mult,m2]-phi_y[n+mult,m2]
            image[n-mult-1,m2]=image[n-mult,m2]-phi_y[n-mult,m2]
    return image

def gaussian_filter(RI):
    rows,columns=RI.shape
    #take fourier transform of image
    ftRI=np.fft.fftshift(RI)
    ft_RI=np.fft.fft2(ftRI)
    ft_RII=np.fft.fftshift(ft_RI)
    ftRI=ft_RII
    #make the 2d gaussian filter
    sigma_x, sigma_y = 55,55
    yi, xi = rows/2, columns/2
    x=np.linspace(0,rows,rows,dtype="float32")
    y=np.linspace(0,columns,columns,dtype="float32")
    X, Y=np.meshgrid(x,y)
    gaussian=np.exp(-(((X-xi)/sigma_x)**2+((Y-yi)/sigma_y)**2))
    
    ftRI_g=ftRI*gaussian
    RI_g=np.abs(np.fft.fftshift((np.fft.ifft2(np.fft.fftshift(ftRI_g)))))
    return RI_g

#call function to store initial phase about the orgin
x_y_phase_map=phase_object(phi_x,phi_y)

#call function to create the full array of x and y phases
phase_x=final_phase_image1(x_y_phase_map,phi_x)
phase_y=final_phase_image2(x_y_phase_map,phi_y)
#average the two phases
phase=(phase_x+phase_y)/2

#read in the images for the suspected binary and it's respective point source
point_acr = cv2.imread('C:\\Users\Emily/hr5676_1a.tif',0)
binary_acr = cv2.imread('C:\\Users\Emily/h075695_1a.tif',0)

#create a mask to get rid of unwanted noise
mask = np.zeros(bispectrum_x.shape, dtype = "float32")
mask=cv2.circle(mask, (128, 128), 55,5, -1)

#find the Fourier Transform of the binary autocorrelation and fix the shift
shift3=np.fft.fftshift(binary_acr)
f_b = np.fft.fft2(shift3)
fshift_b = np.fft.fftshift(f_b)
FT_bin = fshift_b

#find the Fourier Transform of the point source autocorrelation, fix shift
shift4=np.fft.fftshift(point_acr)
f_p = np.fft.fft2(shift4)
fshift_p = np.fft.fftshift(f_p)
FT_point = fshift_p

#divide FT of the binary with the FT of the point to create modulus squared
modulus2 = FT_bin/FT_point
#take the square root to find the modulus
modulus=np.sqrt(modulus2)

#find the imaginary and portion of the object 
object_imag=(modulus)*np.abs(np.sin(phase))
object_real=(modulus)*np.abs(np.cos(phase))

#add the imaginary and real components together
obj = object_imag+object_real
#mask the object with a circle about the center
obj=obj*mask
#apply a gaussian filter after the mask
obj=gaussian_filter(obj)

#inverse fourier transform to find the final reconstructed image  
reconstructed_image = np.abs(np.fft.fftshift((np.fft.ifft2(np.fft.fftshift(obj)))))

#note: orginal image is visually flipped because starting pixel and ending pixel are swapped in y for idl
#note: secondary in this is (118,141), in idl its (118,141) 

#plt.figure(figsize=(5,5))
plt.title('h075695_1a   Reconstructed   Image')
plt.imshow(reconstructed_image,plt.cm.gray)
plt.show()
plt.title('plot of peak of secondary')
plt.plot(reconstructed_image[141])
#plt.axis('off')
plt.show()
plt.title('plot of peak of primary')
plt.plot(reconstructed_image[128,115:141])
plt.show()
#for FWHM in pixels, take max value, divide by 2 then find length
# for delta magnitude, take both plots, find peak and find difference

def mag_diff(RI):
    mask=np.zeros(RI.shape,dtype='float32')
    mask=cv2.circle(mask,(128,128),8,1,-1)
    primary=mask*RI
    primary_peak=np.max(primary)
    secondary=RI-primary
    secondary_peak=np.max(secondary)
    mag_diff_ratio=primary_peak-secondary_peak
    return mag_diff_ratio
#the magnitude difference is a ratio unless primary magnitude is known
magnitude=mag_diff(reconstructed_image)
print('The magnitude difference of the star is:', magnitude)
```
